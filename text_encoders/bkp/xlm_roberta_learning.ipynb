```python
words = {}
for key in d.keys():
    if key not in words:
        words[key] = key
    
    for syn in d[key][1]:
        if syn not in words:
            words[syn] = syn

words_list = list(words.keys())   
```

```python
import numpy as np
import json 

inputs = tokenizer(words_list, return_tensors="pt",padding=True)

# első 100-ra
sub_inputs = {}
sub_inputs["input_ids"] = inputs["input_ids"][0:100]
sub_inputs["attention_mask"] = inputs["attention_mask"][0:100]
print('len(sub_inputs)',len(sub_inputs))
print('len(sub_inputs["input_ids"]',len(sub_inputs["input_ids"]))

outputs = model(**sub_inputs)
print('len(outputs)',len(outputs))

print(outputs.keys())
print('len(outputs.logits)',len(outputs.logits)) # összes elem
print('len(outputs.logits[0])',len(outputs.logits[0])) # első elem
print('len(outputs.logits[0][0])',len(outputs.logits[0][0])) # első elem első rétege
emb=np.mean(outputs.logits.detach().numpy(),axis=1) # elemenként veszi a 30 réteg átlagát
print('len(emb)',len(emb)) # összes elem átlagoltan beágyazva
print('len(emb[0])',len(emb[0])) # első elem átlagoltan beágyazva
```

```python
# measurement,automated, old
from scipy.spatial import distance

sum=0
cnt=0
avg=0

for k,v in d.items():
    baseword=k
    encoded_input_base=tokenizer(baseword,return_tensors='pt')
    output_base = model(**encoded_input_base)
    for e in v[1]:
        synonym=e
        encoded_input_synonym=tokenizer(synonym,return_tensors='pt')
        output_synonym = model(**encoded_input_synonym)

        embedded_base=output_base[0][0][0]
        embedded_synonym=output_synonym[0][0][0]

        dist=distance.cosine(embedded_base.detach().numpy(),embedded_synonym.detach().numpy())
        sum=sum+dist
        cnt=cnt+1
# overall result on synonym-dictionary       
avg=sum/cnt
print(avg)
# xlm roberta large (170m) 0.0025427154258488224   ---   xlm roberta base (63m) 0.002523067917173937
```

```python
# old way, just two words, manually
import numpy as np
hu1 = "fiú"
encoded_input1 = tokenizer(hu1, return_tensors='pt')
print('encoded_input1',encoded_input1)
print("len(encoded_input1['input_ids'])",len(encoded_input1['input_ids']))
output1 = model(**encoded_input1)
emb_hu1=output1[0][0][0]
meaned_hu1=np.mean(output1[0][0].detach().numpy(),axis=1)
last_hu1=output1[0][0][-1]
print('len(output1[0][0])',len(output1[0][0])) # number of layers
hu2 = "férfi"
encoded_input2 = tokenizer(hu2, return_tensors='pt')
output2 = model(**encoded_input2)
emb_hu2=output2[0][0][0]
meaned_hu2=np.mean(output2[0][0].detach().numpy(),axis=1)
last_hu2=output2[0][0][-1]
```

```python
# old way, just two words, manually
from scipy.spatial import distance

print(distance.cosine(emb_hu1.detach().numpy(),emb_hu2.detach().numpy())) #large 0.0007125139236450195 fiu,ferfi  -  base 0.002390563488006592 fiu,ferfi
# the meaned result seems to be worse
print(distance.cosine(meaned_hu1,meaned_hu2)) #base 0.017552435398101807 fiu,ferfi
# the last layer proves to be in the middle
print(distance.cosine(last_hu1.detach().numpy(),last_hu2.detach().numpy())) #base 0.01348799467086792 fiu,ferfi
```
