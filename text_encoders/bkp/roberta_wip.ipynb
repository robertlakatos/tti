```python
from transformers import RobertaTokenizer, RobertaModel
import torch

tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
model = RobertaModel.from_pretrained("roberta-base")
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4, width=200, depth=None, stream=None, compact=False, sort_dicts=False)
```

```python
import json

with open('szinonimak.json', encoding='utf-8') as f:
    d = json.load(f)
```

```python
pp.pprint(d)
```

```python
wordsli = []
for key in d.keys():
    wordsli.append(key)
    for syn in d[key][1]:
        wordsli.append(syn)
pp.pprint(wordsli)
#ez azért jobb, mert lehet benne ismétlődés
```

```python
import numpy as np
import json 

inputs = tokenizer(wordsli, return_tensors="pt",padding=True)

sub_inputs = {}
sub_inputs["input_ids"] = inputs["input_ids"][0:100]
sub_inputs["attention_mask"] = inputs["attention_mask"][0:100]
print('len(sub_inputs)',len(sub_inputs))
print('len(sub_inputs["input_ids"]',len(sub_inputs["input_ids"]))

outputs = model(**sub_inputs)
print('len(outputs)',len(outputs))

print(outputs.keys())
print('len(outputs.last_hidden_state)',len(outputs.last_hidden_state))
print('len(outputs.last_hidden_state[0])',len(outputs.last_hidden_state[0]))
print('len(outputs.last_hidden_state[0][0])',len(outputs.last_hidden_state[0][0]))

embs=np.mean(outputs.last_hidden_state.detach().numpy(),axis=1)
print('len(embs)',len(embs))
print('len(embs[0])',len(embs[0]))

#pp.pprint(embs[0])
#pp.pprint(embs[10])
#pp.pprint(embs[50])
#pp.pprint(embs[99])

part_d_li=[]
cnt=0
for w in wordsli[0:100]:
    part_d_li.append((w,embs[cnt]))
    cnt=cnt+1
print('len(part_d_li)',len(part_d_li))
#pp.pprint(part_d)
```
